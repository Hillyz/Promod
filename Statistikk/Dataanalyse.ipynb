```python
def f(x):
    return 9*x**2

def g(x):
    return -x**3 + 2*x**2

def h(x):
    return 10*x

x_verdier = linspace(-10, 10, 1000)

f_verdier = [f(x) for x in x_verdier]

g_verdier = [g(x) for x in x_verdier]

h_verdier = [h(x) for x in x_verdier]

for x in range(10):
    plot(x_verdier, f_verdier, color='hotpink')
    plot(x_verdier, g_verdier, color='limegreen')
    plot(x_verdier, h_verdier, color='gold')
    
xlabel('x-akse')
ylabel('y-akse')
legend(['smil', 'bakke', 'strek'])
show()
```
